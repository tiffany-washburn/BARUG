## Text Analysis

Using the "tm" package, we can convert the data to a corpus, and clean the data.

```python
%%R
doc <- Corpus(VectorSource(as.character(all_text_set$postingbody)[!is.na(all_text_set$postingbody)]))
## inspect(doc[1]) ## View corpus
## Prepare the corpus
doc <- tm_map(doc, stripWhitespace) ## Strips white space
doc <- tm_map(doc, content_transformer(tolower)) ## Convert to lower case
doc <- tm_map(doc, removeNumbers) ## Removes numbers
doc <- tm_map(doc, removePunctuation) ## Removes punctuation
doc <- tm_map(doc, removeWords, stopwords("english")) ## Removes English stop words
doc <- tm_map(doc, stemDocument) ## Stem document
```

Then we create a document term matrix.

```python
%%R
dtm <- DocumentTermMatrix(doc)
```

We can see what words are most highly correlated with things we care about, with "findAssocs."

### Compute Z scores

```python
%%R
freq_comb$prop <- freq_comb$in_sf/sum(freq_comb$in_sf)
freq_comb$prop2 <- freq_comb$in_nat/sum(freq_comb$in_nat)
freq_comb$z <- (freq_comb$prop - freq_comb$prop2)/((sqrt(((sum(freq_comb$in_sf) * freq_comb$prop) + (sum(freq_comb$in_nat) * freq_comb$prop2))/(sum(freq_comb$in_sf) + sum(freq_comb$in_nat)) * (1 - ((sum(freq_comb$in_sf) * freq_comb$prop) + (sum(freq_comb$in_nat) * freq_comb$prop2))/(sum(freq_comb$in_sf) + sum(freq_comb$in_nat))))) * (sqrt((sum(freq_comb$in_sf) + sum(freq_comb$in_nat))/(sum(freq_comb$in_sf) * sum(freq_comb$in_nat)))))  #z formula
```

### Difference plot - SF vs. National Sample

```python
%%R
## Create visualization
freq_comb <- freq_comb[order(abs(freq_comb$z), decreasing = T), ]  #order according to significance
# calculate percentage reduction:
freq_comb$dif1 <- -100 * (1 - freq_comb$prop/freq_comb$prop2)
# calculate percentage increase
freq_comb$dif2 <- 100 * (1 - freq_comb$prop2/freq_comb$prop)
# merge results. Very clunky, must be better way:
freq_comb$dif <- 0
freq_comb$dif[freq_comb$dif1 < 0] <- freq_comb$dif1[freq_comb$dif1 < 
                                                               0]
freq_comb$dif[freq_comb$dif2 > 0] <- freq_comb$dif2[freq_comb$dif2 > 
                                                               0]
freq_comb$dif1 <- NULL
freq_comb$dif2 <- NULL
require(ggplot2)
ggplot(freq_comb, aes(z, dif)) + geom_point()
```

Note the non-linear relationship between the absolute difference in Z-score.  We're interested in both the size and the significance of differences in the corpi.


```python
%%R -w 960 -h 960 -u px  
freq_comb$z2 <- 1
freq_comb$z2[abs(freq_comb$z)>=1.96] <- 0
freq_comb <- freq_comb[freq_comb$z>-99&freq_comb$z<99,]
#exclude terms infitiely more or less used
freq_comb <- freq_comb[order(abs(freq_comb$prop2+freq_comb$prop),decreasing=T),] 
##freq_comb$in_sf <- freq_comb$in_sf+runif(nrow(freq_comb), -3, 3)
ggplot(head(freq_comb,50), #select number of points in plot
       aes(dif,log(abs(in_sf+in_nat)),label=word,size=(in_sf+in_nat),colour=z2))+ #plot z value against difference
  geom_text(fontface=2,alpha=.8)+
  scale_size(range = c(2, 12))+#specify scale
  ylab("log number of mentions")+
  xlab("percentage difference betwen samples\n <---------More in SF ads ------|-----More in National ads-------->")+
  geom_vline(x=0,colour="red",linetype=2)+theme_bw()+theme(legend.position="none")+
  ggtitle("terms featuring differently in \n San Francisco and National personals ads on Craigslist")

```
